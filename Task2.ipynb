```python
time_column = 'Period'
value_column = 'Sales_quantity'

df[time_column] = pd.to_datetime(df[time_column])

plt.figure(figsize=(15,10))
plt.plot(df[time_column], df[value_column], marker='o', linestyle='-', color='blue')
plt.title(f'Trend Over Time: {value_column}')
plt.xlabel('Time')
plt.ylabel(value_column)
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

```python
time_column = 'Period'
value_column = 'Average_cost'  

# Convert the 'Date' column to datetime if it's not already
df[time_column] = pd.to_datetime(df[time_column])

# Calculate the moving average with a specified window size
window_size = 7  # Adjust the window size based on your preference
df['Moving_Average'] = df[value_column].rolling(window=window_size).mean()

# Plot the original values and the moving average
plt.figure(figsize=(15,10))
plt.plot(df[time_column], df[value_column], label='Original Values', marker='o', linestyle='-', color='blue')
plt.plot(df[time_column], df['Moving_Average'], label=f'Moving Average ({window_size} days)', linestyle='--', color='orange')
plt.title(f'Trend Over Time with Moving Average: {value_column}')
plt.xlabel('Time')
plt.ylabel(value_column)
plt.legend()
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose
```

```python
df.columns
```

```python
df.isnull().sum()
```

```python
df=df.dropna()
```

```python
df
```

```python
columns_to_decompose = ['Revenue', 'Sales_quantity', 'Average_cost', 'The_average_annual_payroll_of_the_region']

# Apply seasonal decomposition for each column
for column in columns_to_decompose:
    result = seasonal_decompose(df[column], model='additive', period=1)  # Adjust the period based on your data

    # Plot the original time series, trend, seasonality, and residuals
    plt.figure(figsize=(12, 8))

    plt.subplot(4, 2, 1)
    plt.plot(df[column], label='Original Time Series')
    plt.legend()

    plt.subplot(4, 2, 2)
    plt.plot(result.trend, label='Trend')
    plt.legend()

    plt.subplot(4, 2, 3)
    plt.plot(result.seasonal, label='Seasonality')
    plt.legend()

    plt.subplot(4, 2, 4)
    plt.plot(result.resid, label='Residuals')
    plt.legend()

    plt.suptitle(f'Seasonal Decomposition of {column}')
    plt.tight_layout(rect=[0, 0, 1, 0.96])
    plt.show()
```
